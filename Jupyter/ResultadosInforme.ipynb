```python
import pandas as pd
```

```python
iters = pd.read_csv('Resultados/tesis_public_datos_iteracion.csv',header=0,names=['id','id_ejec','numero_iteracion','fitness_mejor','fitnessprom','fitnessiter','parametros_iteracion','d1','d2','d3'])
iters.head(1)            
```

```python
iters = pd.merge(iters,tabla[['algorithm','id_ejec','instance']],how='left',on='id_ejec')
```

```python
iters.head()
```

```python
iters['iter'] = iters['numero_iteracion']
iters['fitness'] = iters['fitness_mejor']
```

```python
import json
for idx,rows in iters.iterrows():
    print(idx)
    params = json.loads(rows['parametros_iteracion'])
    
    for key,val in params.items():
        iters.loc[idx,key] = val
    
    break
```

```python
import pandas as pd
import numpy as np
from matplotlib.ticker import FuncFormatter

import matplotlib.pyplot as plt
import math 
plt.style.use('classic')

totalplots = len(tabla.drop_duplicates(['algorithm','instance']).index)
print('==========================================================================================================================')
print(f'Iniciando generaci√≥n de grupos de {totalplots} plots')
print('==========================================================================================================================')

i=0
for alg in tabla.algorithm.unique():
    for inst in tabla[tabla.algorithm==alg].instance.unique():
        i+=1
        
        print(f'[{i}/{totalplots}] Consultando a la db: {alg}-{inst}')
        iters2 = iters[(iters.algorithm==alg)&(iters.instance==inst)]
      
        print(f'[{i}/{totalplots}] Datos ok')

        ax = iters2.pivot(columns='id_ejec',values='fitness',index='iter').plot(figsize=(15,7),legend=False)
        ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.1f}'.format(y))) 
        plt.title(f'Convergence plot for {alg} algorithm solving instance {inst}')
        plt.savefig(f'Resultados/20201126/Convergence_{alg}_{inst}.png',dpi=150)
        plt.savefig(f'Resultados/20201126/eps/Convergence_{alg}_{inst}.eps',dpi=150)
        plt.close()
```
